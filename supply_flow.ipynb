### Hypothesis Testing

To check if there are significant differences in the revenue generated across different customer demographics, we will conduct hypothesis testing.

- **Null Hypothesis**: There is no significant difference in the revenue generated across customer demographics.  
  $$ H_0: \mu_1 = \mu_2 = \mu_3 = \dots = \mu_k $$

- **Alternative Hypothesis (Hâ‚)**: There is a significant difference in the revenue generated across customer demographics.  
  $$ H_1: \text{At least one } \mu_i \neq \mu_0 $$

We will perform a **One-Way ANOVA** to test for overall differences in revenue across customer demographics, and then use **pairwise Z-tests** (post-hoc analysis) to identify which specific demographics are significantly different from each other.

### Two-Tailed Test
The test will be two-tailed, meaning we are testing if the revenue generated is significantly different (either higher or lower) for any of the demographics.


```python
# One-Way ANOVA for Multiple Demographics

# First, we need to calculate the sample mean for each group - male, female, non-binary, and not defined

# sample mean of the male group

male_sample_mean = revenue_generated[revenue_generated["Customer demographics"] == "Male"]["Revenue generated"].mean()
female_sample_mean = revenue_generated[revenue_generated["Customer demographics"] == "Female"]["Revenue generated"].mean()
binary_sample_mean = revenue_generated[revenue_generated["Customer demographics"] == "Non-binary"]["Revenue generated"].mean()
unknown_sample_mean = revenue_generated[revenue_generated["Customer demographics"] == "Unknown"]["Revenue generated"].mean()


```

```python
# Calculate the grand mean
grand_mean = revenue_generated["Revenue generated"].mean()
grand_mean
```

```python
male_sample_variance = revenue_generated[revenue_generated["Customer demographics"] == "Male"]["Revenue generated"].var()
female_sample_variance = revenue_generated[revenue_generated["Customer demographics"] == "Female"]["Revenue generated"].var()
binary_sample_variance = revenue_generated[revenue_generated["Customer demographics"] == "Non-binary"]["Revenue generated"].var()
unknown_sample_variance = revenue_generated[revenue_generated["Customer demographics"] == "Unknown"]["Revenue generated"].var()

n_male = len(revenue_generated[revenue_generated["Customer demographics"] == "Male"])
n_female = len(revenue_generated[revenue_generated["Customer demographics"] == "Female"])
n_non_binary = len(revenue_generated[revenue_generated["Customer demographics"] == "Non-binary"])
n_unknown = len(revenue_generated[revenue_generated["Customer demographics"] == "Unknown"])


# Finding the Between-group Sum of Squares (SSB)
SSB = (n_male * (male_sample_mean - grand_mean)**2 +
       n_female * (female_sample_mean - grand_mean)**2 +
       n_non_binary * (binary_sample_mean - grand_mean)**2 +
       n_unknown * (unknown_sample_mean - grand_mean)**2)

print(f"The Between-group Sum of Squares value comes out to be {SSB}")
```

```python
# Calculating the Within-Group Sum of Squares 
SSW_male = ((revenue_generated[revenue_generated["Customer demographics"] == "Male"]["Revenue generated"] - male_sample_mean)**2).sum()
SSW_female = ((revenue_generated[revenue_generated["Customer demographics"] == "Female"]["Revenue generated"] - female_sample_mean)**2).sum()
SSW_non_binary = ((revenue_generated[revenue_generated["Customer demographics"] == "Non-binary"]["Revenue generated"] - binary_sample_mean)**2).sum()
SSW_unknown = ((revenue_generated[revenue_generated["Customer demographics"] == "Unknown"]["Revenue generated"] - unknown_sample_mean) **2).sum()

SSW = SSW_male + SSW_female + SSW_non_binary + SSW_unknown

print(f"The Within-Group Sum of Squares value comes out to be {SSW}")
```

```python
SST = ((revenue_generated["Revenue generated"] - grand_mean)**2).sum()
print(f"The total sum of square values comes out to be {SST}")
```

```python
#VERIFICATION 
SST_verify = SSW + SSB 
SST_verify
```

```python
# Find the F-Ratio
dof_bw = 3 # Number of groups is 4 and dof is k - 1, where k is the number of groups
dof_within = len(revenue_generated) - 4

MSB = SSB/dof_bw
MSW = SSW/dof_within
```

```python
F_ratio = MSB/MSW
print(f"F-Ratio: {F_ratio}")
```

```python
from scipy.stats import f

p_value = 1 - f.cdf(F_ratio, dfn=dof_bw, dfd=dof_within)
print(f"P-Value: {p_value}")
```

Conclusion: There is no significant evidence to suggest that the revenue generated differs across customer demographics. In simpler terms, the differences in revenue means across the demographics (Male, Female, Non-binary, Unknown) are likely due to random variation and is not statistically significant.
