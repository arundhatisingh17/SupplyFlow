```python
# Task 1: My main objective here is to analyze the reenue generated by each product type broken down by the customer demographics

revenue_generated = supply_chain_data.groupby(["Product type", "Customer demographics", "Location"])["Revenue generated"].sum()
revenue_generated = revenue_generated.reset_index()
revenue_generated
```

```python
revenue_table = revenue_generated.pivot_table(
    index = ["Product type", "Customer demographics"],
    columns = "Location",
    values = "Revenue generated",
    aggfunc = "sum"
)


revenue_table
```

```python
import folium
india_map = folium.Map(location=[20.5937, 78.9629], zoom_start=5)
india_map2 = folium.Map(location=[20.5937, 78.9629], zoom_start=5)
india_map
```

```python
revenue_generated[['Latitude', 'Longitude']] = revenue_generated['Location'].apply(lambda x: pd.Series(get_lat_lon(x)))

# To assign colors based on the customer demographics
def get_color(demographic):
    if demographic == "Female":
        return 'pink'
    elif demographic == "Male":
        return 'blue'
    elif demographic == "Non-binary":
        return 'purple'
    else:
        return 'gray'
    

for _, row in revenue_generated.iterrows():
    folium.CircleMarker(
        location = [row["Latitude"], row['Longitude']],
        radius = 15,
        color = get_color(row["Customer demographics"]),
        fill = True,
        fill_color = get_color(row["Customer demographics"]),
        fill_opacity = 0.4,
    ).add_to(india_map)

india_map
```

```python
from folium.plugins import HeatMap

# Prepare data for heatmap
heat_data = [[row["Latitude"], row["Longitude"], row["Revenue generated"]] for _, row in revenue_generated.iterrows()]

# Add HeatMap to the map
HeatMap(heat_data).add_to(india_map)

india_map
```
